# Sales Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sales = pd.read_csv("sales_data.csv")
#columns : Product_ID,	Sale_Date,	Sales_Rep,	Region,	Sales_Amount,	Quantity_Sold,	Product_Category,	Unit_Cost,	Unit_Price,	Customer_Type,	Discount,	Payment_Method,	Sales_Channel,	Region_and_Sales_Rep

font1={'family':'serif', 'color':'blue', 'size':20}
font2={'family':'serif', 'color':'red', 'size':18}
font3={'family':'serif', 'color':'green', 'size':18}


sales["Total_sales"]= sales['Quantity_Sold'] * sales['Unit_Price']

sns.barplot(data=sales, x='Product_Category', y="Total_sales", hue="Product_Category")
plt.title("Total Sales by Product", fontdict=font1)
plt.xlabel("Product Category", fontdict=font3)
plt.ylabel("Total Sales (€)", fontdict=font2)
plt.show()
```

```python
sns.barplot(data=sales, x='Total_sales', y='Payment_Method', hue='Payment_Method')
plt.title("Total Sales by Payment Method", fontdict=font1)
plt.xlabel("Total Sales (€)", fontdict=font3)
plt.ylabel("Payment Method", fontdict=font2)
plt.show()
```

```python
sales_by_region = sales.groupby('Region')['Total_sales'].sum()

plt.figure(figsize=(6,6))
plt.pie(sales_by_region, labels=sales_by_region.index, autopct='%1.1f%%')
plt.title("Sales Distribution by Region", fontdict=font1)
plt.show()

```

```python
sns.boxplot(data=sales, x="Total_sales", y="Sales_Channel", hue="Sales_Channel")
plt.title("Total Sales by Sales Channel", fontdict=font1)
plt.xlabel("Total Sales (€)", fontdict=font3)
plt.ylabel("Sales Channel", fontdict=font2)
```

Conclusions:


*   The best-selling product category is clothing and the least-selling is food.
*   The most used payment method is Bank Transfer and the least used is Credit card.



*   The region with the most sales was the north and the region with the least sales was the south.
*   More online sales than retail.
